# exploratory data analysis and data cleaning

```python
df.isnull().sum()
```

```python
#creating new column age group
df.loc[df["Age"]>=50,"age group"]="Senior" 
df.loc[df["Age"]<50,"age group"]="Adult" 
df.loc[df["Age"]<30,"age group"]="Teenger"
df.head()
```

```python
df['age group'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
#repaling W, M with women and men
df['Gender'] = df['Gender'].replace(["M","W"],["Men","Women"])
df
```

```python
df['Qty'].value_counts()
```

```python
#replacing One and Two with 1,2
df['Qty'] = df['Qty'].replace(["One","Two"],[1,2])
df
```

```python
# creating new month column
df['month'] = df['Date'].dt.month_name()
df
```

```python
df.to_excel("data analysis.xlsx",index=False)
```
